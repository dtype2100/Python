```python
data_list = []
for data_csv in data:
  data_df = pd.read_csv(data_csv, index_col=0)
  data_list.append(data_df)

data_concat_df = pd.concat(data_list)
print('num', len(data_concat_df))
print(data_concat_df.head())
```

```python
data_df = [] # df를 담을 리스트를 준비
journal_df = []

for data_csv in data:
    df = pd.read_csv(data_csv, encoding='utf-8')
    df['year'] = data_csv[-8:-4]
    data_df.append(df)
    
len(data_df)
concat_df = pd.concat(data_df)
```
