```python
n_df = naver_df[['상호', '별점', '음식분류', '주소', '지번', '전화번호', '위도', '경도', 'urls']]
m_df = mango_df[['title', 'rating', 'new_category', 'category', 'menu', 'address', 'tel', 'lat', 'lng', 'url']]
k_df = kakao_df[['stores', 'rating', 'category', 'menu', 'road_address', 'phone_number', 'Y', 'X', 'place_url']]
```

```python
n_df.columns
```

```python
n_df.rename(columns={'상호': 'n_상호', '별점':'n_별점', '음식분류': 'n_음식', '주소':'n_주소', '지번':'n_지번', '전화번호':'n_전화번호', '위도':'n_위도', '경도':'n_경도', 'urls': 'n_url'}, inplace=True)
```

```python
m_df.rename(columns={'title':'m_상호', 'rating':'m_별점', 'new_category':'m_대분류', 'category':'m_음식', 'menu':'m_메뉴', 'address':'m_지번', 'tel':'m_전화번호', 'lat':'m_위도', 'lng':'m_경도', 'url':'m_url'}, inplace=True)
```

```python
k_df.rename(columns={'stores':'k_상호', 'rating':'k_별점', 'category':'k_음식', 'menu':'k_메뉴', 'road_address':'k_지번', 'phone_number':'k_전화번호', 'Y':'k_위도', 'X':'k_경도', 'place_url':'k_url'}, inplace=True)
```

```python
k_m_df = pd.merge(k_df, m_df, how='left', left_on= 'k_전화번호', right_on='m_전화번호')
```

```python
k_m_n_df = pd.merge(k_m_df, n_df, how='left', left_on= 'k_전화번호', right_on='n_전화번호')
```
