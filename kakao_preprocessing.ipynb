```python
encoder = LabelEncoder()
```

```python
encoder.fit(docs)
labels = encoder.transform(docs)
```

```python
#labels = labels.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_scale = scaler.fit_transform(labels.reshape(-1,1))
```

```python
km = KMeans(n_clusters=20).fit(data_scale.reshape(-1,1))
```

```python
df['labels'] = labels
df['scaler'] = data_scale
df['km.labels_'] = km.labels_
```
