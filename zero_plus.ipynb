# 시간 앞에 0 추가하기

```python
import pandas as pd
```

```python
data = pd.DataFrame({'A': [0, 0, 0],
                     'B': [900, 900, 900],
                     'C': [1500, 1500, 1500]})
data
```

```python
data['D'] = ['0' + str(x) for x in data['A'] if len(str(x)) == 3 or len(str(x)) == 1]
data['E'] = ['0' + str(x) for x in data['B'] if len(str(x)) == 3 or len(str(x)) == 1]
data['F'] = ['0' + str(x) if len(str(x)) == 3 or len(str(x)) == 1 else x for x in data['C']]
```
